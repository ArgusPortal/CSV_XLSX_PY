```python
import os
import pandas as pd

def csv_to_xlsx(input_folder, output_folder):
    # Verifica se a pasta de saída existe, se não, a cria
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Lista todos os arquivos na pasta de entrada
    files = os.listdir(input_folder)

    for file in files:
        if file.endswith(".csv"):
            csv_path = os.path.join(input_folder, file)

            # Lê o arquivo CSV com o pandas, especificando a codificação
            df = pd.read_csv(csv_path, encoding='utf8')

            # Gera o nome do arquivo XLSX na pasta de saída
            xlsx_file = os.path.splitext(file)[0] + ".xlsx"
            xlsx_path = os.path.join(output_folder, xlsx_file)

            # Escreve o DataFrame no arquivo XLSX
            df.to_excel(xlsx_path, index=False)

            print(f'Conversão concluída: {file} -> {xlsx_file}')

# Substitua "/content/caminho/da/pasta/csv" pelo caminho da pasta que contém os arquivos CSV no seu ambiente Colab
input_folder = "/content/"
# Substitua "/content/caminho/da/pasta/xlsx" pelo caminho da pasta onde os arquivos XLSX serão salvos no seu ambiente Colab
output_folder = "/content/XLSX"

csv_to_xlsx(input_folder, output_folder)

```

```python
# BAIXAR TODA A PASTA ZIPADA

import shutil

# Compacta a pasta de saída em um arquivo zip
shutil.make_archive("/content/XLSX", 'zip', "/content/XLSX")

# Move o arquivo zip para /content para facilitar o download
shutil.move("/content/XLSX.zip", "/content/XLSX.zip")
```
